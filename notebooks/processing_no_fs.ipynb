```python
import numpy as np
import pandas as pd

import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
URL = "../notebooks/pre_processed.csv"
games_original = pd.read_csv(URL, encoding='UTF-8')
games = games_original
```

```python
games[['owners_min', 'owners_max']] = games['owners'].str.split('-', expand=True)
games['owners'] = games['owners_min']
games = games.drop(['owners_min','owners_max'], axis = 1)
games['owners'] = pd.to_numeric(games['owners'], errors='coerce')
```

```python
games.drop("achievements", axis=1, inplace=True)
games.iloc[::, 0:18].head(20)
```

```python
games.to_csv('features_selected.csv', index=False)  
```
